```python
file_name = 'whd_whisard.naicHumanReadableLevels.csv'
column_1 ='cty_nm'
column_2 = 'bw_atp_amt'
wage_code = pd.read_csv(file_name)

mean = []
maximum = []
employees = []
company = []
for description in wage_list_city:
    section = wage_code[wage_code[column_1]==description] 
    section_bw = section[column_2]
    mean.append(np.mean(section_bw))
    maximum.append(max(section_bw))
    employees.append(sum(section['ee_violtd_cnt']))
    company.append(section['trade_nm'].iloc[np.argmax(list(section[column_2]))])

    
    
```

```python
back_pay_city['Mean'] = mean
back_pay_city['Employees'] = employees
back_pay_city['Max'] = maximum
back_pay_city['Company'] = company
```

```python
company_v2 = []
for maximum, company in zip(back_pay_city['Max'],back_pay_city['Company']):
    if maximum == 0.00:
        company = "N/A"
        company_v2.append(company)
    else:
        company_v2.append(company)
```

```python
back_pay_city['Company'] = company_v2
```
